```python
import pandas as pd
import numpy as np
import faiss
import time

from sentence_transformers import SentenceTransformer

import google.generativeai as genai

from google.colab import userdata
```

```python
GOOGLE_API_KEY = userdata.get(
    "Research++"
)

genai.configure(
    api_key=GOOGLE_API_KEY
)

llm = genai.GenerativeModel(
    "gemini-2.5-flash"
)

print(
    "Gemini configured successfully."
)
```

```python
df = pd.read_json(
    "papers.json"
)

faiss_index = faiss.read_index(
    "researchforge_faiss.index"
)

print(
    f"Loaded {len(df)} papers."
)
```

```python
embedding_model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)

print(
    "Embedding model ready."
)
```

```python
def retrieve_papers(
    user_query,
    top_k=10
):

    query_embedding = embedding_model.encode(
        [user_query]
    )

    distances, indices = faiss_index.search(
        np.array(
            query_embedding
        ).astype("float32"),
        k=top_k
    )

    retrieved_papers = []

    for idx in indices[0]:

        retrieved_papers.append({

            "title":
            df.iloc[idx]["title"],

            "abstract":
            df.iloc[idx]["abstract"]

        })

    return retrieved_papers
```

```python
def run_research_debate(
    user_query,
    retrieved_papers
):

    evidence_context = ""

    for paper in retrieved_papers:

        evidence_context += f"""

Title:
{paper['title']}

Abstract:
{paper['abstract']}

"""

    prompt = f"""

You are simulating a scientific debate.

Research Question:

{user_query}

Evidence:

{evidence_context}

Create:

Agent A:
Supports one position.

Agent B:
Challenges that position.

Both agents must cite
evidence from the papers.

Then create:

Judge Agent:

Compare both arguments.

Identify strongest evidence.

Identify weaknesses.

Provide final conclusion.

Keep discussion scientific,
balanced and evidence grounded.

"""

    for attempt in range(3):

        try:

            response = llm.generate_content(
                prompt
            )

            return response.text

        except Exception:

            print(
                f"Retry {attempt+1}/3 ..."
            )

            time.sleep(60)

    return (
        "Debate generation failed."
    )
```

```python
query = """
Are transformer-based multimodal
deepfake detectors better than
traditional multimodal approaches?
"""

papers = retrieve_papers(
    query
)

debate_output = run_research_debate(
    query,
    papers
)

print(
    debate_output
)
```
